## Sevinj caluclation but for genomics

```python
def Sevinj_method_class(clinical, features_df):
    path='../../Data/'

    ch_cl=pd.read_csv(path + clinical + '.csv')
    chum_gen=pd.read_csv(path + features_df + '.csv')
    chum_gen = helper.prep_genomics(chum_gen)
    genetic_features = chum_gen.columns[1:]
    new_d=pd.merge(ch_cl,chum_gen, on='oncotech_id')
    print(len(new_d))
    a=new_d.os_days.quantile(0.333)
    b=new_d.os_days.quantile(0.666)
    c=new_d.os_days.quantile(0.666)
    print("Q1 quantile: ", new_d.os_days.quantile(0.33))
    print("Q2 quantile: ", new_d.os_days.quantile(0.66))
    print("Q3 quantile: ", new_d.os_days.quantile(0.66))
    print(a,b,c)
    df1=new_d[new_d.os_days<=a].reset_index()
    df2=new_d[(new_d.os_days>a) & (new_d.os_days<=b)].reset_index()
    df3=new_d[new_d.os_days>=b].reset_index()

    print(len(df1), len(df3))
    column_name='class_os_days'

    gen_first = df1
    gen_second = df3
    gen_first[column_name] = 0
    gen_second[column_name] = 1
    gen_chum=pd.concat([gen_first, gen_second], ignore_index=True)
    gen_chum=gen_chum.sample(frac=1).reset_index(drop=True)
    return gen_chum, genetic_features
```

```python
gen_chum, genetic_features = Sevinj_method_class("clinical_CHUM", "Discovery-GenomicsTop5000-March22023")
gen_chum[list(genetic_features)+["class_os_days", "oncotech_id"]].to_csv("../../Data/CHUM_genetics_class.csv")
```

```python
gen_iucpq, genetic_features = Sevinj_method_class("clinical_IUCPQ", "ValidationIUCPQ-March2 2023")
gen_iucpq[list(genetic_features)+["class_os_days", "oncotech_id"]].to_csv("../../Data/IUCPQ_genetics_class.csv")
```

```python
def add_os_day_class(df, predictor):
    df["class_"+ predictor] = pd.qcut(df[predictor], 3, labels=False, retbins=False, precision=3, duplicates='raise')
    df = df[df["class_"+ predictor] != 1]
    df["class_"+ predictor][df["class_"+ predictor] == 2] = 1
    df["class_"+ predictor][df["class_"+ predictor] == "nan"] = np.nan
    df = df[df["class_"+ predictor].notna()]
    
    df = df.reset_index(drop=True)
    #df = df.drop(columns=[predictor, "Unnamed: 0"])
    df["class_"+ predictor] = df["class_"+ predictor].astype('int')
    return df
```
