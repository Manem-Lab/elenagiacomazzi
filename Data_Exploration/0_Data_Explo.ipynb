## Gene Expression

```python
geneExp_chum = pd.read_csv("../../Data/GeneExpression_CHUM.csv")
geneExp_chum.rename(columns={"Unnamed: 0": "Protein"}, inplace=True)

geneExp_iucpq = pd.read_csv("../../Data/GeneExpression_IUCPQ.csv")
geneExp_iucpq.rename(columns={"Unnamed: 0": "Protein"}, inplace=True)

```

```python
def hard_facts(df, print_it=True):
    nr_proteins = len(df.Protein.unique())
    nr_patients = len(df.columns[(df.columns.str.endswith("CHUM")) | (df.columns.str.endswith("IUCPQ"))])
    if print_it:
        #print(df.columns[(df.columns.str.endswith("CHUM")) | (df.columns.str.endswith("IUCPQ"))])
        print("{} unique proteins exist in the data.".format(nr_proteins))
        print("{} unique patients exist in the data.".format(nr_patients))

print("CHUM")
hard_facts(geneExp_chum)
print("IUCPQ")
hard_facts(geneExp_iucpq)
```
